```python
Train_data.loc[(Train_data['Married/Single'] == 'single'),"Married/Single"] = 0
Train_data.loc[(Train_data['Married/Single'] == 'married'),"Married/Single"] = 1
Train_data["Married/Single"] = Train_data["Married/Single"].astype(str).astype(int)
Train_data.loc[(Train_data['Car_Ownership'] == 'no'),"Car_Ownership"] = 0
Train_data.loc[(Train_data['Car_Ownership'] == 'yes'),"Car_Ownership"] = 1
Train_data["Car_Ownership"] = Train_data["Car_Ownership"].astype(str).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

rng = np.random.RandomState(42)

# Generate train data
X = df.iloc[:,0:6]
y=df.iloc[:,6]
X_train, X_test, Y_train, Y_test =train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
Y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score
print(f'Accuracy={accuracy_score(Y_test,Y_predict):.2f}\nRecall={recall_score(Y_test,Y_predict):.2f}\n')
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predict)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,cmap="Blues")
plt.xlabel('Predict')
plt.ylabel('Truth')
```
